```python
import pandas as pd
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
   
```

```python
plt.plot(k_values, accuracy_scores, marker="o")
plt.title("Accuracy Score vs K Value")
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
plt.xticks(k_values)
plt.grid(True)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("fruit_data_with_colors _1_.csv")
df = df.drop(["fruit_name", "fruit_subtype"], axis=1)

imputer = SimpleImputer(strategy="mean")
df_imputer = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)
```

```python
X = df_imputer.drop("fruit_label", axis=1)
y= df_imputer["fruit_label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
k_values = range(1,11)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy = knn.score(X_test, y_test)
    accuracy_scores.append(accuracy)

accuracy_scores
```

```python
plt.plot(k_values, accuracy_scores, marker="o")
plt.title("K Values vs Accuracy Score")
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
plt.xticks(k_values)
plt.grid(True)

```
